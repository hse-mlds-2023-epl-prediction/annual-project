# EDA для таблиц matches, players, teams

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
players = pd.read_csv('players.csv')
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.histplot(players['Age'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

numeric_columns = ['Age', 'Starts', 'MathesPlayed', 'Minutes', 'Goals', 'Assists', 'PenaltyAttempt',
                    'YellowCards', 'RedCards', 'xG', 'xAG', 'ProgressivePasses', 'ProgressiveCarriers']
plt.figure(figsize=(12, 8))
sns.heatmap(players[numeric_columns].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='MathesPlayed', y='Goals', data=players, hue='Position', palette='viridis')
plt.title('Scatter plot of Matches Played vs. Goals with Position Separation')
plt.xlabel('Matches Played')
plt.ylabel('Goals')
plt.legend(title='Position', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(players['YellowCards'], bins=20, kde=True, color='gold')
plt.title('Distribution of Yellow Cards')
plt.xlabel('Number of Yellow Cards')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(players['RedCards'], bins=20, kde=True, color='red')
plt.title('Distribution of Red Cards')
plt.xlabel('Number of Red Cards')
plt.ylabel('Frequency')
plt.show()
```

```python
teams = pd.read_csv('teams.csv')

def str_to_float(value):
    if pd.notna(value): 
        return float(str(value).replace(',', '.'))
    else:
        return value

teams['AttendanceHomePerGame']  = teams['AttendanceHomePerGame'].apply(str_to_float)

sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.histplot(teams['AttendanceHomePerGame'].dropna(), bins=20, kde=True, color='orange')
plt.title('Distribution of Attendance Per Game at Home')
plt.xlabel('Attendance Per Game')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Wins', y='Losses', data=teams, hue='Season', palette='viridis')
plt.title('Scatter plot of Wins vs. Losses')
plt.xlabel('Wins')
plt.ylabel('Losses')
plt.legend(title='Season', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

```

```python
matches = pd.read_csv('matches.csv')
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.histplot(matches['Attendance'].dropna(), bins=20, kde=True, color='blue')
plt.title('Distribution of Match Attendance')
plt.xlabel('Attendance')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(12, 6))
sns.histplot(matches[['HomeTeamYellowCardsCount', 'AwayTeamYellowCardsCount']], bins=20, kde=True, color=['green', 'red'])
plt.title('Distribution of Yellow Cards for Home and Away Teams')
plt.xlabel('Yellow Cards')
plt.ylabel('Frequency')
plt.legend(labels=['Home Team', 'Away Team'])
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='HomeTeamPossessionPercent', y='HomeTeamShotsTotalCount', data=matches, color='purple')
plt.title('Scatter plot of Possession vs. Total Shots for Home Team')
plt.xlabel('Possession Percent')
plt.ylabel('Total Shots')
plt.show()
```
