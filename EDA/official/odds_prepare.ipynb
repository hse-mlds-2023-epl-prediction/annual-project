```python
import pandas as pd
from pathlib import Path
from datetime import datetime
```

```python
path = '../../parsers/official/data/club_stat.csv'
df = pd.read_csv(Path(path).resolve())

path_odds = '../../parsers/odds_portal/matches_odds.csv'
df_odds = pd.read_csv(Path(path_odds).resolve())
```

```python
orig_name = df['name'].sort_values().unique().tolist()
len(orig_name)
```

```python
orig_name 
```

```python
odds_name = df_odds['home_name'].sort_values().unique().tolist()
len(odds_name)
```

```python
map_dict = {}
for i, j in zip(odds_name, orig_name):
    map_dict[i] = j

map_dict
```

```python
df_odds.head()
```

```python
df_odds['home_name'] = df_odds['home_name'].apply(lambda x: map_dict[x])
df_odds['away_name'] = df_odds['away_name'].apply(lambda x: map_dict[x])
df_odds['season'] = df_odds['season'].apply(lambda x: x[:4])
```

```python
df_odds['date'] = df_odds['date_start_timestamp'].apply(lambda x: datetime.utcfromtimestamp(x).date())
df_odds['date'] = pd.to_datetime(df_odds['date'])
```
