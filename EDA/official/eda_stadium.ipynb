```python
import pandas as pd
import seaborn as sns
import re
from pathlib import Path
```

```python
def coordinatesDeg(A):   
    res = re.findall(r'[0-9]+',A)
    res =[int(i) for i in res]
    n = int(len(res)/2)
        
    lat, lon = 0, 0
    for i in range(n) :
        lat += res[i] / (60 ** (i%3))
        lon += res[i+n] / (60 ** (i%3))
        
    return(lat, lon)        
 
```
